```python
import tensorflow as tf
from tensorflow import keras 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Load the dataset
data = keras.datasets.fashion_mnist
```

```python
# Split datset into train and test datasets
(train_images, train_labels), (test_images, test_labels) = data.load_data()
```

```python
# True labels
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Scale train_images and test_images values between 0 and 1
train_images = train_images / 255.0
test_images = test_images  / 255.0
```

## Creating the model

One key concept is to flatten your data turn 3d,2d nd into 1d data input essentialy remove interior lists and pass data as a single list.

```python
# Flatten : data is passed as a single list in the input layer
# Dense : every neurone is connected to every neurone in the next layer (fully connected layer)
# Sequential : define layers in order, input then deep layer then output.
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
# our output layers is 10 neurones because we have 10 labels
# our input layer will be 28*28 = 784 neurones
# our deep layer will be 128 neurones it's a prct of the input layer neurones number
```

```python
# compiling the model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
# Train the model
model.fit(train_images, train_labels, epochs=10)
```

```python
# Test the model
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
```

### Using the model for predictions

```python
predictions = model.predict(test_images)
predictions[0]
```

```python
# Use the real class name and return one value instead of 10
class_names[np.argmax(predictions[0])]
```

```python
# Validate the predictions
for i in range(10):
    plt.grid(False)
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    plt.xlabel("Actual: " + class_names[test_labels[i]])
    plt.title("Model Prediction: " + class_names[np.argmax(predictions[i])])
    plt.show()
```

```python
# Predict on one image
prediction = model.predict(np.expand_dims(test_images[7], axis=0))
print("Predicted: " + class_names[np.argmax(prediction)])
print("Actual: " + class_names[test_labels[7]])
```
